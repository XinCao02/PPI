```python
input_file_path = "3did_flat.txt"
output_file_path = "3did_filtered.txt"

with open(input_file_path, "r") as infile, open(output_file_path, "w") as outfile:
    for line in infile:
        if line.startswith("#=3D"):
            outfile.write(line)

print(f"Filtered lines have been saved to {output_file_path}.")
```

```python
filtered_file_path = "3did_filtered.txt"  
unique_pdb_file_path = "unique_pdb_ids.txt"  

pdb_ids = set()

with open(filtered_file_path, "r") as filtered_file:
    for line in filtered_file:
        fields = line.split()  # 按空格分割
        pdb_id = fields[1].upper()  # 提取第二列并转换为大写
        pdb_ids.add(pdb_id)  # 添加到集合中去重

with open(unique_pdb_file_path, "w") as unique_file:
    for pdb_id in sorted(pdb_ids):  # 排序后写入文件
        unique_file.write(pdb_id + "\n")

print(f"Unique PDB IDs have been saved to {unique_pdb_file_path}.")
```

```python
# Define the file name
input_file = "Sample_GCN/human_PPI_string_20240505.txt"

# Initialize counters
col5_zero_count = 0
col5_nonzero_count = 0
col6_zero_count = 0
col6_nonzero_count = 0
both_zero_count = 0
both_nonzero_count = 0
anom_ct = 0

# Open and process the file
with open(input_file, "r") as file:
    # Skip the header line
    header = file.readline()
    

    # Process each line
    for line in file:
        try:
            # Split the line into columns
            columns = line.strip().split("\t")
            
            # Extract columns 5 and 6 (index 4 and 5 in zero-based indexing)
            col5 = int(columns[4])
            col6 = int(columns[5])
            
            # Count zeros and nonzeros in col5
            if col5 == 0:
                col5_zero_count += 1
            else:
                col5_nonzero_count += 1
            
            # Count zeros and nonzeros in col6
            if col6 == 0:
                col6_zero_count += 1
            else:
                col6_nonzero_count += 1
            
            # Count Both col5 and col6 are 0
            if col5 == 0 and col6 == 0:
                both_zero_count += 1
            
            # Count Both col5 and col6 are nonzero
            if col5 != 0 and col6 != 0:
                both_nonzero_count += 1
        except:
            anom_ct += 1

# Print the results
print(f'Anom Count: {anom_ct:,}')
print(f"Column 5 - Zero count: {col5_zero_count:,}")
print(f"Column 5 - Nonzero count: {col5_nonzero_count:,}")
print(f"Column 6 - Zero count: {col6_zero_count:,}")
print(f"Column 6 - Nonzero count: {col6_nonzero_count:,}")
print(f"Lines where both col5 and col6 are zero: {both_zero_count:,}")
print(f"Lines where both col5 and col6 are nonzero: {both_nonzero_count:,}")
```
